#### **üßê Why is Natural Language Processing a challenge?**
Human language is rich, dynamic and layered with complexities that are deeply intuitive for humans but difficult for machines to grasp. Some of the key challenges include:

**a. Ambiguties**

Well, language is nothing if not complicated. Three striking ambiguities emerge:
> **1. Lexical ambiguity:** when a word has more than one possible meaning and the meaning has to be inferred from the context in which the word is used. For example;
- ***She went to get gold from the bank***. Which bank?  A river bank, or an actual bank?
> **2. Syntactical ambiguity:** when a phrase has more than one potnetial meaning. For example:
- ***Visiting relatives can be annoying***
> **3. Referential ambiguity:** occurs when we use pronouns. For example:
- ***It was she who started the confrontation with him***. Who do we refer to as him? Who is she?

These ambiguities make writing down deterministic algorithms for language very difficult. Equally, the level of accuracy required in any language tasks cannot simply be achieved by taking a heuristic approach. 

**b. Dynamic nature**

Language is ever evolving, with new additions to the vocabulary, inflected forms, flexible boundaries that allow the introdcution of words from other languages to another language. 

**c. Nuances**

Language is characterised by nuances such as slang, irony, sacarsm that are learned through use of the language and perceived from context. This additional layer of complexity further contributes to an already daunting task of making computers understand human language.

```python
def pre_process_text(corpus):
    """
    Applies the following pre-processing steps to text:
        - lowercasing
        - removing punctuation
        - removing stopwords
        - tokenisation
        - lemmatisation
    Params:
        - corpus(text): a multiline string of the text to be pre-processed
    Returns:
        - offspring(list): an nested array containing pre-processed tokens from each sentence
    """
    offspring = []
    word_lemmatizer = WordNetLemmatizer()
    # creating sentences
    sentences = sent_tokenize(corpus)
    stopwords_en = set(stopwords.words("english"))
    for sent in sentences:
        item = re.sub(r"[^a-zA-Z]",' ',sent)      # removing punctuation and numeric characters
        item = item.lower() # making lowercase
        item = item.split() # word tokenisation
        item = [word for word in item if word not in stopwords_en] # filtering out stopwords using list comprehension
        item = [word_lemmatizer.lemmatize(token) for token in item] # base forms of the words
        offspring.append(item)
    return offspring
```

#### **üî¢ Bag of Words** ***(What words appear?)***

**Idea behind Bag of Words:** Any language, has a finite vocabulary/no. of words. However, in NLP we deal with long passages, texts and paragraphs. For any one of these, we certainly will not use the entire vocabulary of that language. Think about it, not unless you crawl the depth of the internet and find every book, article or text ever written, it's very hard to come up with a text that contains all the words in the English vocabulary. Having this in mind, what other way can we numerically represent words in a specific text?

One way is to treat all unique words in that document as the entirerity of the vocabulary of the language. Therefore, we can capture the presence or absence of a word in a particular sentence or phrase. Each sentence can then be represented as a vector and the entire document in turn as a collection of vectors. This collection of unique words is what we call a **bag of words**

##### **üõ†Ô∏è Building a bag of words model from scratch**

Once we have our 'bag of words', we represent each document as a vector. The length of this vector equals the total number of unique words in the text(the size of our vocabulary). Each position in our vectors hold the count of how many times the corresponding words appears in that particular document. 
##### **‚úçüèæ Demonstration**
Let's build the BoW model:

```python
docs_df
```

```python
# creating a dataframe bag of words
bow_df = pd.DataFrame()
bow_df['docs'] = docs_df['doc']
# adding the tokens present in each sentence
bow_df['tokens'] = docs_df['tokens']
bow_df.head()
```

```python
# creating columns for each unique token
for col in sorted(vocab):
    bow_df[col] = 0
# creating the bag of words representation
for i,tokens in enumerate(bow_df['tokens']):
    for token in tokens:
        if token in bow_df.columns:
            bow_df.loc[i,token]=tokens.count(token) # the value is the number of times that feature occurs in that sentence
# dropping the token column
bow_df.drop(columns=['tokens'],inplace=True)
bow_df
```

The dataframe above shows the bag of words representation of the text. Here's what our text now looks like with the numerical representation's of each document.

```python
vector_cols = list(bow_df.columns)
vector_cols.remove('docs')
assert len(vector_cols)==112
```

```python
numerical_text = []
for i in range(len(bow_df)):
    numerical_text.append(list(bow_df.loc[i,vector_cols].values))

for doc, nums in zip(docs_df['doc'],numerical_text):
    print(f"Actual document: {doc}")
    print(f"BoW representation: {nums}")
```

```python
def clean_mail(email_text):
    """
    Text pre-processing for email text
    Params:
        - email_text(pd.Series): a series containing email text to be processed
    Returns:
        - emails_clean(list): an array containing processed entries from the pandas Series
    """
    emails_clean = []

    for mail in email_text:
        new_mail = re.sub(r"[^a-zA-Z0-9$', ]","", mail)
        mail_tokens = word_tokenize(new_mail)

        tagged_tokens = pos_tag(mail_tokens)

        clean_lemmas = []
        for token,tag in tagged_tokens:
            pos = lemmatizer_map.get(tag)
            lemma = lemmatizer.lemmatize(token, pos) if pos else lemmatizer.lemmatize(token)
            lemma_clean = re.sub(r"[^a-z]","",lemma.lower())
            if lemma_clean not in en_stopwords:
                clean_lemmas.append(lemma_clean)
        emails_clean.append(re.sub(r'\s+', ' ',(" ".join(clean_lemmas)).strip()))
    return emails_clean
```

##### **üîÆFrom Frequency to Importance - What comes after Bag of Words**
While Bag of Words is a good starting point for representing text data in a nunerical format, language is nuanced and thus we need to encode more than simply the occurence and frequency of a word. We need other approaches that build up from the bag of words model. We can attempt to encode the semantics of a word by assigning weights to each word in a document. The higher the weight, the more meaning or value to the text the word carries. This can be done using a technique known as ***TF-IDF***.

#### **üìàüíéTF-IDF**
**Term Frequency Inverse Document Frequency(TF-IDF)** is used to encode text data by assigning weights to tokens based on two factors:
- How frequently a word appears in a document (Term Frequency)
- How rare that word is across all documents (Inverse Document Frequency)

This means that TF-IDF increases the weight of words that are important to a specific document but uncommon across the entire corpus. The weight of a token is determined as:

$$

w_{ij} = tf_{ij} \cdot \ln\left(\frac{N}{df_i}\right)

$$

Where:
- $( w_{ij})$ is the TF-IDF weight for term \( i \) in document \( j \).
- $( tf_{ij})$ is the term frequency of term \( i \) in document \( j \).
- $( N)$ is the total number of documents.
- $( df_i)$ is the document frequency of term \( i \) (number of documents containing term \( i \)).

$$
tf_{ij} = \frac{\text{Number of repetitions of token } i \text{ in sentence } j}{\text{Total number of tokens in sentence }j}
$$
$$
idf_i = \log \left( \frac{\text{Total number of sentences}}{\text{Number of sentences containing token } i} \right)
$$

##### **üìå Why Are the Weights Different?**
From the demonstration above, we can clearly see that the weights of tokens in the TF-IDF model we built from scratch are different from the those built using the TfIdf vectroizer class from the scikit-learn library. 

Well if the underlying formula is the same, **ü§∑üèø‚Äç‚ôÇÔ∏èWhy the difference?**

The TF-IDF implementation in the scikit-learn libary is a bit different from that described by the formula.

**üëÅÔ∏èDifference between the two approaches**

- Count matrix for the term frequency part(similar to a Bag of Words)
- Smoothed inverse document frequency by:
    - Adding one to the document count
    - Adding one to the count of documents containing token i
    - Adding one to the final result of the inverse document frequency part
- Normalisation of the final weights

**üìê scikit-learns' TF-IDF formula**

The weight of a token is now determined as:

$$
tf_{ij} = \text{Number of occurences of token } i \text{ in sentence } j
$$

$$
idf_i = \log \left( \frac{\text{Total number of sentences + 1}}{\text{Number of sentences containing token } i + 1} \right)
$$
The final weight:
$$

w_{ij} = tf_{ij} \cdot (\ln\left(\frac{N}{df_i}\right) + 1)

$$

$$
w^{norm}_{ij} = \frac{w_{ij}}  {||w_{j}||}
$$
where $||w_{j}||$ is the l2 norming constant for document vector j

**ü§îWhy does the TF-IDF class in sci-kit learn use this approach?**

When algorithms are incorporated into standard libraries, they need to work reliably across diverse applications and data conditions without compromising performance or usability.

Since scikit-learn is primarily a machine learning library, it makes these design choices for practical reasons:

1. Smoothing the IDF prevents division-by-zero errors and overly extreme values when terms appear in very few or very many documents.
2. Normalization ensures that resulting TF-IDF vectors are on a common scale, which:

    - Accelerates model convergence during training.
    - Enables fair distance or similarity comparisons (especially for metrics like cosine similarity).
    - Mirrors best practices like feature scaling in numeric ML models.

This makes scikit-learn‚Äôs TF-IDF both robust and ML-friendly by design.

### **N-grams**

We have seen how the `Bag of Words` models and `TF-IDF` models work. However, there is a key limitation to this approach, while having a numerical representation of each word is great, we fail to capture on of the glaring complexities of language, **dependency**. For instance ***"not good"*** and ***'good'*** mean different things. The meaning of phrases is affected by the phrases quite often around them or even a few words apart. This is a key feature of language that needs to be captured. 

One innovative approach to this is to select combinations of tokens rather than single tokens and then encode those numerically; instead of having features `['The','quick','brown','fox','jumped','over','the','fence]` for the sentence ***"The quick brown fox jumped over the fence"*** we would have say `['The quick','quick brown','brown fox',...]`. In this way we can capture the relationships between words too.

How long the combinations are is given by **N** with common algorithms such as **Bi-grams** for ***N=2***. The BagofWords model is therefore a special case of the N-grams approach with ***N=1***.

### **Word2Vec**

The underlying nuance is that most of the data available in the world is unstructured. One of the forms that unstructured data that has powered the collaboration in advancements of the human race is language(text in our case). It is therefore important to come up with approaches to not only extract insight from text but also attempt to understand it programatically. Why? Well machines are better at crunching numbers than we are and we could leverage this. So any form of data, we can effectively get the most out of it if we can encode it in numerical format. 

Previously, we have used `Bag of Words` and `TF_DF` to encode text in numerical form. However, these two approaches come with one shared pitfall, ***they fail to capture the semantics of a word(meaning)***. When I came accross this weakness, I thought to myslef, well, why not simply encode the dictionary in numerical format and let the vector representations of a token in a document interact with that from the dictionary and this will capture the meaning. However, this has a problem; the model created will be good at mathcing a token to its lexical meaning(what is defined in a dictionary) kind of like a lookup table. However, language is quite unfortunately for us more complex than that. See language is made powerful by the unique combination of words to make up phrases and sentences. However, we do not uunderstand the meaning of a phrase or a sentence based of its lexicon, we infer it from the context in which the words have been used, e.g: `The bank was so shallow` can be expressed in lexicon as: ``['article','the bottom of a river','past tense of is','expression of extent' and 'not deep']`` however from the simple sentence we immediately know that `We are referring to a river bank and that it was shallow`. This can only be achieved from context. Therefore, the ideal go to approach would be to focus on capturing **semantic meaning** rather than lexial meaning of language.

So we have established that there key thing in decoding the semantic meaning of text:
    - `The context in which it has been used` e.g I went to the bank and I got some gold from the river bank 

Therefore, we need an approach to encoding text that captures contextual relationships between words. One such revolutionary approach is **Word2Vec**. Word2Vec has the following architecture:
    - You have a corpora of text from a document(a corpora here is simply all the words present in a document). Some pre-processing tasks might have been applied say tokenisation and removing stop words.
    - You create a high dimensional vector referred to as an **embedding** say 32 or 64 dimensions.
    - The high dimensional vector becomes the unique embedding of that word.


But how can we do this?

We can do this by considering the position in which a word occurs in the text, what is emphasised in transformers through **positional encoding** and ensuring that we process text in the order in which it occurs therefore for each word used in a different context we have a different numerical embedding for it. 

But yeah great but how do we really do this?

I know of 2 approaches native to the word2vec model. A very genius way of determining the context is attempting to predict the next word or a the next group words. There are 2 approaches for this:

1. `Convolutionary Bag of Words(CBOW)`: convolutionary bag of words attempts to predict the next word given some words e.g: For the sentence "The quick brown fox jumped over the lazy dog" we extract a list of words such as ['the','quick','brown','fox','jumped','over','the','lazy','dog']. From this list we take n words and attempt to predict what the next word will be:

    - say we have n=3, so we condier 3 words in sequence and try and predict the next word we'd have ['the','quick','brown'] to predict ['fox'] and so on hence the name convolutionary bag of words.
    This is repated over and over again creating a sequence of features and targets. 

2. `Skip-gram`: in this approach we attempt to predict the next series of words/tokens given a single word. So ['quick] to predict say ['brown','fox'].

The first approach sound easier and more natural but both have their benefits. SKip-gram is especially good for encoding rare words.

So we have approaches to predicting the next word how does that get us closer to numerical representation of the words?

So the engine powering all this is yet another algorithm known as a neural network. An interconnected network of nodes known as neurons. The word embeddings come from the weights of the neural network. The final input weights to the hidden layer after the network is trained. This is hopw it works:

1. Each token in the corpora is converted into a numerical vector of dimension N say 7(depends on how many unique tokens are present) using one-hot encoding so basically a token such as `fox` would be [0,0,0,0,1,0,1]. We do this for all the tokens where the one marks their presence.
2. Some of the token(s) are considered features and another a target and passed to a single input node of a neural network. Attempting to predict the target from the features. A neural netweork will typically have multiple input nodes.
3. Weights are initialised randomly for each node as connections to the hidden layer. 
4. The firt layer of the hidden layer is very crucial, from these the dimension of our word emedding is determined e.g for 32 dimension embedding the first node needs to have 32 neurons for 32 weights for each input node. These weights are what will serve as our input embedding after training is complete. 
5. The training loop is initiated and after an acceptable loss if attained, the final callibrated weights to the hidden layer are used as word embeddings for each target feature.

Now execution of this seems conplex but is quite simplified by the word2vec model from gensim. Neural networks are known for being notoriously compute hungry. However, the neural network used here is typically a very shallow network.
So what happens when you call the word2vec model and pass it your document is the above process is carried out. There are some nuances such as loss optimisation and activations that I have not described for the neural net which we can have some form of control over using the parameters available from the word2Vec model.

Now we can then do some interesting things with the embeddings. For instance, how close 2 word embeddings are tells us how similar in meaning two words are and the further apart words are from each other the less likely they are related.

```python
def extract_text(web_urls):
    """
    Extracts text content from wikipedia pages into one corpus

    Args:
        - web_urls(list): array containing all the web addresses to obtain data from

    Returns:
        corpus (list): array containing all the text from all websites
        topics (list): array highlighting the topics scraped for
    """
    corpus = []
    topics = []
    for url in web_urls:
        topic = url.split("/")[-1]
        topics.append(topic)
        print(f"Extracting wikipedia text -> {topic} üîÉüîÉ")
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        paragraphs = soup.find_all('p')
        for p in paragraphs:
            corpus.append(p.get_text())

    return corpus, topics
```

```python
def make_skipgrams(token_sequences, window_size = 2):
    skipgrams = []
    for seq in token_sequences:
        for token in range(len(seq)):

            context_idx = [i for i in range(token - window_size, token + window_size + 1) if i!=token and 0<= i < len(seq)]

            for cont in context_idx:
                skipgrams.append((seq[token], seq[cont]))

    return skipgrams
```

```python
def make_targets_features(skipgram_seq):
    features = []
    targets = []

    for comb in skipgram_seq:
        features.append(comb[0])
        targets.append(comb[-1])

    return features, targets
```
